11) Plot um histograma com a quantidade de domicilios desta base por UF.

```python
plt.figure(figsize=(10,10))
ordem = ["Cômodo", "Apartamento", "Casa"]

ax1 = sns.histplot(x="uf", data=df, hue="tipo_domicilio", hue_order=ordem, alpha=0.75, palette = "tab10", kde=True)
ax1.set_title('Quantidade de Domicílios por UF', size=24)
ax1.set_xlabel('UF')
ax1.set_ylabel('Quantidade de Domicílios')
ax1.get_legend().set_title('Tipos de Domicílios')

plt.xticks(rotation=90)
plt.show()
```

12) Plot um gráfico de barras com a soma total de moradores desta base por UF.

```python
df2 = df[["uf",'num_moradores']].groupby("uf").agg(Soma_Total = ('num_moradores',"sum"))
df2.reset_index(inplace=True)
df2.sort_values('Soma_Total', ascending=False, inplace=True)
df2.head()
```

```python
plt.figure(figsize=(10,10))

ax2 = sns.barplot(x="uf", y="Soma_Total", data=df2, palette ="Greens_r")
ax2.set_title('Soma Total de Moradores por UF', size=24)
ax2.set_xlabel('UF')
ax2.set_ylabel('Soma Total de Moradores')

plt.xticks(rotation=90)
plt.show()
```

13) Plot um gráfico de dispersão com o número total de moradores (num_moradores) e o número total de dormitorios (num_dormitorios).

```python
df3 = df[["uf","num_moradores","num_dormitorios"]].groupby("uf").agg(total_moradores=('num_moradores','sum'),
                                                                     total_dormitorios=('num_dormitorios','sum')
                                                                     )
df3.reset_index(inplace=True)
df3.head()
```

```python
plt.figure(figsize=(10,10))

ax3 = sns.scatterplot(x="total_moradores",y="total_dormitorios", data=df3, s=200, color="green")
ax3.set_title('Número Total de Moradores x Número Total de Dormitórios', size=16)
ax3.set_xlabel('Total de Moradores')
ax3.set_ylabel('Total de Dormitórios')

plt.show()
```
